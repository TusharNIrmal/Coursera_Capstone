### DEFINED A FUNCTION  "get_coords_local" .IT RETURNS COORDINATES OF THE GIVEN LOCALITY .  IT HAS 2 PARAMETERS - 1.) locality - NAME OF THE LOCALITY WE WANT COORDINATED OF . 2.) "output_as" - TO CHOOSE FROM ONE OF THE TWO OPTIONS - 'boundingbox' OR 'center' .  . 

```python
def get_coords_local(locality, output_as='center'):
    """
    get the bounding box of a locality in WGS84 given its name

    Parameters
    ----------
    locality : str
        name of the country in english and lowercase
    output_as : 'str
        chose from 'boundingbox' or 'center'. 
         - 'boundingbox' for [latmin, latmax, lonmin, lonmax]
         - 'center' for [latcenter, loncenter]

    Returns
    -------
    output : list
        list with coordinates as str
    """
    # create url
    url = '{0}{1}{2}'.format('http://nominatim.openstreetmap.org/search.php?q=',
                             locality +', Pune , India',
                             '&format=json&polygon=0')
    if len(requests.get(url).json()) == 0:
        output = ['Not Found','Not Found']
        return output
        
    response = requests.get(url).json()[0]

    # parse response to list
    if output_as == 'boundingbox':
        lst = response[output_as]
        output = [float(i) for i in lst]
    if output_as == 'center':
        lst = [response.get(key) for key in ['lat','lon']]
        output = [float(i) for i in lst]
    return output
```

### WE MAKE A COPY OF "nb_pune_df" AS NAME IT AS "nb_pn_coor" . IN THIS DATAFRAME  , WE ADD 2 MORE COLUMNS - "Latitude" AND "Longitude" . WE CALL THE FUNCTION "get_coords_local" AND PASS THE NAMES OF EACH LOCALITY ONE BY ONE THROUGH "for" LOOP AND APPEND THE COORDINATES INTO NEW LISTS. THE VALUES FOR THESE COLUMNS ARE THEN OBTAINED BY ASSIGNING EACH COLUMN WITH ITS CORRESPONDING LIST . 

### THUS AT THE END , OUR DATAFRAME "nb_pn_coor" CONTAINS THREE COLUMNS - "Localities" , "Latitude", "Longitude" .

```python
nb_pn_coor = nb_pune_df.copy()

latitudeCln = []
longitudeCln = []
for index, row in nb_pn_coor.iterrows():
    print(row[0])
    lat, long = get_coords_local(locality=row[0], output_as='center')
    latitudeCln.append(lat)
    longitudeCln.append(long)

nb_pn_coor['Latitude'] = latitudeCln
nb_pn_coor['Longitude'] = longitudeCln

nb_pn_coor.shape
```

```python
# function to exploring venues of all the neighborhoods in Pune

def getNearbyVenues(names, latitudes, longitudes, radius=5000, categoryIds=''):
    try:
        venues_list=[]
        for name, lat, lng in zip(names, latitudes, longitudes):
            #print(name)

            # create the API request URL
            url = 'https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, radius, LIMIT)

            if (categoryIds != ''):
                url = url + '&categoryId={}'
                url = url.format(categoryIds)

            # make the GET request
            response = requests.get(url).json()
            results = response["response"]['venues']

            # return only relevant information for each nearby venue
            for v in results:
                success = False
                try:
                    category = v['categories'][0]['name']
                    success = True
                except:
                    pass

                if success:
                    venues_list.append([(
                        name, 
                        lat, 
                        lng, 
                        v['name'], 
                        v['location']['lat'], 
                        v['location']['lng'],
                        v['categories'][0]['name']
                    )])

        nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
        nearby_venues.columns = ['Localities', 
                  'Locality Latitude', 
                  'Locality Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    except:
        print(url)
        print(response)
        print(results)
        print(nearby_venues)

    return(nearby_venues)
```

### defining  a function to columns to the dataframe 

```python
def addColumn(startDf, columnTitle, dataDf):
    grouped = dataDf.groupby('Localities').count()
    
    for n in startDf['Localities']:
        try:
            startDf.loc[startDf['Localities'] == n,columnTitle] = grouped.loc[n, 'Venue']
        except:
            startDf.loc[startDf['Localities'] == n,columnTitle] = 0
```
